```python
from collections import Counter

# Create a function to calculate the percentage of shared words between two lists
def shared_word_percentage(list1, list2):
    count1 = Counter(list1)
    count2 = Counter(list2)
    common = sum((count1 & count2).values())
    total = sum(count1.values())
    return common / total if total > 0 else 0

# Create a list of pairs with their corresponding shared word percentages
pairs_with_percentages = []
for i in range(min(len(list1), len(list2))):
    percentage = shared_word_percentage(list1[i], list2[i])
    pairs_with_percentages.append((i, percentage))

# Sort the pairs by percentage in descending order
sorted_pairs = sorted(pairs_with_percentages, key=lambda x: x[1], reverse=True)

# Get the top 50 pair indexes
top_50_pair_indexes = [pair[0] for pair in sorted_pairs[:50]]

print("Top 50 Pair Indexes:", top_50_pair_indexes)
```

```python
incons_idx = []
words = []
contr = []


for i, j in zip(df.iterrows(), IG_words.iterrows()):

    vals = j[1]['label']
    dat = j[1]['words']

    og_data = i[1]['Sentences'].replace('\u200c', '').split()



    start_idx = 0

    og_idx = 0

    fin_vals = []

    for y in range(len(dat)):
        if og_idx >= len(og_data):
            incons_idx.append((i[0], 'og idx mismatch'))
            break
        if len(''.join(dat[start_idx:y+1])) >= len(og_data[og_idx]):

            if (''.join(dat[start_idx:y+1])) == og_data[og_idx]:
                if sum(vals[start_idx:y+1]) > 0:
                    fin_vals.append(1)
                else:
                    fin_vals.append(0)

                start_idx = y+1
                og_idx += 1

            elif (''.join(dat[start_idx:y+1])).strip() == og_data[og_idx]:

                if sum(vals[start_idx:y+1]) > 0:
                    fin_vals.append(1)
                else:
                    fin_vals.append(0)

                start_idx = y+1
                og_idx += 1

            elif(''.join(dat[start_idx:y+1])).strip() == og_data[og_idx].strip():
                if sum(vals[start_idx:y+1]) > 0:
                    fin_vals.append(1)
                else:
                    fin_vals.append(0)

                start_idx = y+1
                og_idx += 1



            elif set((''.join(dat[start_idx:y+1])).strip()) == set(og_data[og_idx].strip()) and '\u200c' not in set((''.join(dat[start_idx:y+1])).strip()) and '\u200c' not in set(og_data[og_idx].strip()):

                if sum(vals[start_idx:y+1]) > 0:
                    fin_vals.append(1)
                else:
                    fin_vals.append(0)

                start_idx = y+1
                og_idx += 1

            elif (len(''.join(dat[start_idx:y+1])) >= len(og_data[og_idx])) and og_data[og_idx] in ''.join(dat[start_idx:y+1]):
                if sum(vals[start_idx:y+1]) > 0:
                    fin_vals.append(1)
                else:
                    fin_vals.append(0)

                start_idx = y+1
                og_idx += 1
            else:
                t1 = [i for i in set((''.join(dat[start_idx:y+1])).strip()) if i != '\u200c']
                t2 = [i for i in set(og_data[og_idx].strip()) if i != '\u200c']

                if sorted(t1) == sorted(t2):
                    if sum(vals[start_idx:y+1]) > 0:
                        fin_vals.append(1)
                    else:
                        fin_vals.append(0)

                    start_idx = y+1
                    og_idx += 1



    if len(fin_vals) != len(og_data):
        incons_idx.append((i[0], 'len mismatch'))
        fin_vals += [0] * (len(og_data) - len(fin_vals))
    else:
        words.extend(og_data)
        contr.extend(fin_vals)

    words.extend('.')
    contr.extend('2')
```

```python
import re
import numpy as np

def merge_tokens(tokens, scores, text):

    merged_words = []
    merged_word_scores = []

    current_word = ''
    current_score = 0

    word_pattern = r'\w+'

    scores = np.array(scores)

    for match in re.finditer(word_pattern, text):
        word = match.group()
        length = len(word)

        while length > 0 and tokens.size > 0:
            token = tokens[0]
            score = scores[0]

            if token in ['[CLS]', '[SEP]']:
                tokens = tokens[1:]
                scores = scores[1:]
                continue

            token = token.replace("##", "")

            current_word += token
            current_score += score
            length -= len(token)

            tokens = tokens[1:]
            scores = scores[1:]

        merged_words.append(current_word)
        merged_word_scores.append(current_score)

        current_word = ''
        current_score = 0

    return merged_words, merged_word_scores
```

```python
def print_org(sentence, human_annotated):
  for word in sentence.split():
    if word in human_annotated:
      print(f'\033[1;37;42m {word}' , end="")
    else:
      print(f'\033[0;30;47m {word}' , end="")

def print_model(sentence, true_preds, false_preds):
  # print("\033[0;30;47m Correct: ", true_preds, "Wrong: ", false_preds)
  for word in sentence.split():
    if word in true_preds:
      print(f'\033[1;37;42m {word}' , end="")
    elif word in false_preds:
      print(f'\033[1;37;41m {word}' , end="")
    else:
      print(f'\033[0;30;47m {word}' , end="")
```

```python
import ast

sentences = finbert['Sentences']
human_words = finbert['Human_words']
finbert_words = finbert['Model_Words']
roberta_words = roberta ['Model_Words']
higlish_words = hinglish['Model_Words']

for sent, word, fin, rob, hing in zip(sentences, human_words, finbert_words, roberta_words, higlish_words):

  word = ast.literal_eval(word)
  fin = ast.literal_eval(fin)
  rob = ast.literal_eval(rob)
  hing = ast.literal_eval(hing)

  fin_true_preds = list(set(fin).intersection(word))
  fin_false_preds = list(set(fin) - set(word))

  rob_true_preds = list(set(rob).intersection(word))
  rob_false_preds = list(set(rob) - set(word))

  hing_true_preds = list(set(hing).intersection(word))
  hing_false_preds = list(set(hing) - set(word))

  print("\033[0;30;47m Original Sentence: ", end="")
  print_org(sent, word)
  print('\n')

  print("\033[0;30;47m FinBert: ", end="")
  print_model(sent, fin_true_preds, fin_false_preds)
  print('\n')

  print("\033[0;30;47m Hinglish: ", end="")
  print_model(sent, hing_true_preds, hing_false_preds)
  print('\n')

  print("\033[0;30;47m Roberta: ", end="")
  print_model(sent, rob_true_preds, rob_false_preds)
  print('\n')

  print('\n')
  print('\n')
```
